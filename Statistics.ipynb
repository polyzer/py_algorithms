## Confidence interval

```python
import numpy as np
```

```python
sample = np.random.randint(0, 400, 1000)
sample
```

```python
average = np.sum(sample)/len(sample)
average
```

```python
R = np.max(sample) - np.min(sample)
R
```

```python
(sample -average)**2
```

```python
disp = np.sum((sample - average)**2)/(len(sample)-1)
disp
```

```python
mean_quad = np.sqrt(disp)
mean_quad
```
